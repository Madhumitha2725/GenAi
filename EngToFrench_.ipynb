```python
from transformers import MarianMTModel, MarianTokenizer

# Load model and tokenizer for English to French translation
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Get user input
text = input("Enter a sentence in English to translate into French: ")

# Tokenize and translate
inputs = tokenizer(text, return_tensors="pt", padding=True)
translated = model.generate(**inputs)

# Decode and print the result
print("Translated sentence:", tokenizer.decode(translated[0], skip_special_tokens=True))
```
